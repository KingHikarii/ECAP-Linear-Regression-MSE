**Linear Regression**

Linear Regression is statistical model that was found in the field of statistics. It assumes that there is a linear relationship between the input variables and the output variables.

It is a linear equation where the input variables are denoted as x, and the output variable is denoted as y. It assumes that specified set of combination of the input variables 'x' will give a predicted output which is known as 'y'.

It has a one-scale factor which is denoted as 'B', and this is a coefficient within our equation, there is also an additional coefficient which is also denoted as B, and this will give our line a degree of freedom, this is called as bias or intercept.

To learn the linear regression algorithm our goal is to estimate the values of our coefficients which in the representation of our data. In machine learning, when we are talking about learning a model, we are saying we are training a math model in order for it to make predictions. In linear regression, what will be doing is estimating the values of our coefficient known as B0 and Bn, in order to minimize the difference between the actual and also the predicted values.

**The Algorithm**

I. **Develop the Gradient Function**
  1. Initialize the coefficient m_gradient and b_gradient as zero.
  2. Create a for loop that will run as many time as the give iterations.
  3. Call in the data points for the independent variable (X) and the dependent variable (Y).
  4. Update m_gradient and b_gradient by using the gradient function.
  5. Create the variables m_now and b_now that will have the new values, initialize this by multipling the current m_gradient and b_gradient by the learning rate.
  6. Return m_now and b_now

II. **Within the Global Function**
  1. Get the data.
  2. Create and initialize the variables m and b to 0. Establish the learning rate and the number of iterations.
  3. Toss those variables and the data into the gradient descent function.
  4. Get the data points of x and y.
  5. Receive the predicted values of y_pred
  5. Plot the data points of the original data.
  6. Plot the new predicted points.
  7. Calculate the mean square error.

**Establishing the Gradient Descent**

1. Initialize the values of each coeffient to 0.
2. Calculate the predicted output using the current coefficients.
3. Calculate the error of the gradient descent.
4. Use the learning rate as a scale factor as the coefficient are going in the direction that minimizes the errors.
5. Process is repeated until a local minimum is found.

```python
def gradient_descent(m_now, b_now, points, L, iterations):
    n = len(points)

    for _ in range(iterations):
        m_gradient = 0
        b_gradient = 0

        for i in range(n):
            x = points.iloc[i].Hours
            y = points.iloc[i].Scores
            m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
            b_gradient += -(2/n) * (y - (m_now * x + b_now))

        m_now -= m_gradient * L
        b_now -= b_gradient * L

    return m_now, b_now
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data points
points = pd.read_csv('/content/score.csv')

# Initial parameters
m_now = 0
b_now = 0
L = 0.01  # Learning rate
iterations = 1000  # Number of iterations

# Perform gradient descent
m, b = gradient_descent(m_now, b_now, points, L, iterations)
print(m, b)

# Predicted values
X = points['Hours']
Y = points['Scores']
Y_pred = m * X + b
print(Y_pred.info())

# Plotting the data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Plotting the trend line
plt.plot(X, Y_pred, color='red', label='Trend Line')

# Adding labels and title
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Trend Line using Gradient Descent')
plt.legend()

# Showing the plot
plt.show()
```

**Mean Squared Error**

*Mean squared error* is metric that is used to determine how far the predicted value is from the true underlying values. If the mean square error's value is high than there isn't much accuracy in our model, and the predicted and the true values are far from each other. It measures the average squared difference between the predicted values and the true values.

```python
import numpy as np
def mean_squared_error(y_pred, Y):
    return np.mean((y_pred - Y) ** 2)

mse = mean_squared_error(Y_pred, Y)
print(mse)

rsme = np.sqrt(mse)
print(rsme)
```

```python
# If y is a NumPy array or pandas Series
print(Y.min())    # minimum value
print(Y.max())    # maximum value
print(Y.mean())   # mean value
print(Y.std())    # standard deviation

y_mean = Y.mean()
```

```python
def cv(rsme,y_mean):
  return (rsme/y_mean)*100

print(cv(rsme, y_mean))
```

```python
def mean_absolute_error(y_pred, Y):
  return np.mean(np.mean(np.abs((y_pred - Y))))

print(mean_absolute_error(Y_pred, Y))
```
